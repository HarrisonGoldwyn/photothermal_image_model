```python
## Get the dndt array (as 2d function of temperature and frequency)
hws = fit_dndT['hws']
dndt_array = fit_dndT['dndt_array']
dnr_dt_array = dndt_array.real
## -------------------

## Calculate finite difference
dni_dt_array = dndt_array.imag
## -------------------
```

```python
a1 = 10
# eps_back = 1.473**2
eps_back = 1.473**2
```

```python
def get_self_interf_terms_vs_energy(T_expansion, r_th, w0, phi, ax=None):
    """ Lengths are input in nm
        """

    ## Get dielectric function and dn/dT
    fit_drude = cp.drude_lorentz_model(
            hws/hbar, 
            *temp_dep_eps_fits[f'{T_expansion:d}K']['x']
            )
    fit_dndT = dndt_array[:, T_expansion]

    (
        alpha0_term, 
        alphapi2_term, 
        alphaphi_term, 
        alphapi2phi_term
        ) = pti.ptsignal_scattering_self_interf_term(
            eps_c=fit_drude,
            eps_b=eps_back,
            a=a1*1e-7,
            b=w0*1e-7,
            r_th=r_th*1e-7,
            dndT_c= fit_dndT,
            phi=phi,
#             P_abs=None,
#             dndT_b=dndT_glycerol,
#             kappa=kappa_glycerol,
#             E_probe=1,
#             G_=1,
            return_components=True,
            )
    return alpha0_term, alphapi2_term, alphaphi_term, alphapi2phi_term

def plot_self_interf_terms_vs_energy(T_expansion, r_th, w0, phi, ax=None):
    """ Lengths are input in nm
        """
    if ax is None:
        fig, ax = plt.subplots(1, 1)
    ## Get dielectric function and dn/dT
    fit_drude = cp.drude_lorentz_model(
            hws/hbar, 
            *temp_dep_eps_fits[f'{T_expansion:d}K']['x']
            )
    fit_dndT = dndt_array[:, T_expansion]

    (
        alpha0_term, 
        alphapi2_term, 
        alphaphi_term, 
        alphapi2phi_term
        ) = get_self_interf_terms_vs_energy(T_expansion, r_th, w0, phi)
    
    ax.plot(
        hws.real,
        alpha0_term,
        label=r"$\alpha_\Omega^{(0)}$")
    ax.plot(
        hws.real,
        alphapi2_term,
        label=r"$\alpha_\Omega^{\pi/2}$",
        ls='--')
    ax.plot(
        hws.real,
        alphaphi_term,
        label=r"$\alpha_\Omega^{(\phi_{w_0})}$")
    ax.plot(
        hws.real,
        alphapi2phi_term,
        label=r"$\alpha_\Omega^{(\pi/2+\phi_{w_0})}$",
        ls='--')
    ax.plot(
        hws.real,
        (  
            alpha0_term
            +alphapi2_term
            +alphaphi_term
            +alphapi2phi_term),
        label=r"sum",
        c='k')
    ax.legend(loc='center left')
    ax.set_ylabel(r'Photo thermal signal, self interf. ')
    ax.set_xlabel(r'$\hbar\omega$')
    plt.suptitle(r'$w_0 =$'+f' {round(w0):d} [nm] '+r'$\phi = $'+f' {phi:.2f}')
    
    return ax


def get_probe_interf_terms_vs_energy(
    T_expansion, r_th, w0, phi, 
    V=1,
    ax=None,
    isolate_comp=None
    ):
    """ Lengths are input in nm
        """

    ## Get dielectric function and dn/dT
    fit_drude = cp.drude_lorentz_model(
            hws/hbar, 
            *temp_dep_eps_fits[f'{T_expansion:d}K']['x']
            )
    fit_dndT = dndt_array[:, T_expansion]

    (
        alpha0_term, 
        alphapi2_term, 
        alphaphi_term, 
        alphapi2phi_term
        ) = pti.ptsignal_probe_scat_interf_term(
            eps_c=fit_drude,
            eps_b=eps_back,
            a=a1*1e-7,
            b=w0*1e-7,
            r_th=r_th*1e-7,
            dndT_c= fit_dndT,
            phi=phi,
#             P_abs=None,
#             dndT_b=dndT_glycerol,
#             kappa=kappa_glycerol,
            V=V,
#             G_=1,
            return_components=True,
            )
    return alpha0_term, alphapi2_term, alphaphi_term, alphapi2phi_term


def plot_probe_interf_terms_vs_energy(
    T_expansion, r_th, w0, phi, 
    V=1,
    ax=None,
    isolate_comp=None
    ):
    """ Lengths are input in nm
        """
    if ax is None:
        fig, ax = plt.subplots(1, 1)

    (
        alpha0_term, 
        alphapi2_term, 
        alphaphi_term, 
        alphapi2phi_term
        ) = get_probe_interf_terms_vs_energy(
            T_expansion, r_th, w0, phi, 
            V=V,
            ax=ax,
            isolate_comp=isolate_comp
            )
    if (isolate_comp == 0) or (isolate_comp is None):
        ax.plot(
            hws.real,
            alpha0_term,
            label=r"$\alpha_\Omega^{(0)}$")
    if (isolate_comp == 1) or (isolate_comp is None):
        ax.plot(
            hws.real,
            alphapi2_term,
            label=r"$\alpha_\Omega^{\pi/2}$",
            ls='--')
    if (isolate_comp == 2) or (isolate_comp is None):
        ax.plot(
            hws.real,
            alphaphi_term,
            label=r"$\alpha_\Omega^{(\phi_{w_0})}$")
    if (isolate_comp == 3) or (isolate_comp is None):
        ax.plot(
            hws.real,
            alphapi2phi_term,
            label=r"$\alpha_\Omega^{(\pi/2+\phi_{w_0})}$",
            ls='--')
    if (isolate_comp is 'total') or (isolate_comp is None):
        ax.plot(
            hws.real,
            (  
                alpha0_term
                +alphapi2_term
                +alphaphi_term
                +alphapi2phi_term),
            label=r"sum",
            c='k')
    ax.legend(loc='center left')
    ax.set_ylabel(r'Photo thermal signal, probe interf. ')
    ax.set_xlabel(r'$\hbar\omega$')
    plt.suptitle(r'$w_0 =$'+f' {round(w0):d} [nm] '+r'$\phi = $'+f' {phi:.2f}')
    
    return ax
```

### Calculation of the effective scattering volume for the transmitted probe field

```python
def vol_x_Einc(hw, eps_b, probe_power=2e-4):
    
    k = (hw * eps_b**0.5 / hbar) / c

    V_times_Einc = np.sqrt(
        (3 * probe_power)
        /
        (c * k**4.)
        )
    return V_times_Einc
    
```

We have the hypothesis that the Claire term dominates for systems like the nanorod dimer. Can I push the model there to support this?

```python
def get_both_PTI_terms_vs_phase(T_expansion, r_th, w0, hw, phis=None):
    """ Returns probe-interference, self-interference PTI terms and their sum. 
        Internally computed effective scattering volume for probe-interf 
        term.
        """
    if phis is None:
        phis = np.linspace(0, 2*np.pi, 100)
    (
        phis, 
        probe_interf_0, 
        probe_interf_1, 
        probe_interf_2, 
        probe_interf_3
        ) = get_probe_interf_terms_vs_phase(
            T_expansion, 
            r_th, 
            w0, 
            hw, 
            phis=phis,
            V=vol_x_Einc(hw=hw, eps_b=eps_back)
            )
    probe_interf = (
        probe_interf_0 + probe_interf_1 + probe_interf_2 + probe_interf_3)
    
    (
        phis, 
        self_interf_0, 
        self_interf_1, 
        self_interf_2, 
        self_interf_3
        ) = get_self_interf_terms_vs_phase(
            T_expansion, 
            r_th, 
            w0, 
            hw,
            phis=phis)
    
    self_interf = (
        self_interf_0 + self_interf_1 + self_interf_2 + self_interf_3)
    
    ## Combine to get the total PTI signal
    total = probe_interf + self_interf
    
    max_phi = phis[np.argmax(total)]
    print(f'Signal maximized at phase '+r'$\phi = '+f' {max_phi}')
    
    return phis, probe_interf, self_interf, total
```

```python
phis, test_probe_interf, test_self_interf, test_total = get_both_PTI_terms_vs_phase(
    T_expansion=0, r_th=2000, w0=500, hw=1.8)
```
