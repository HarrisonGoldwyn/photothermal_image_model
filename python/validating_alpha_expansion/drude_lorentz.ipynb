## Try Drude Lorentz Model

```python
fit_cutoff = 2.

def drude_lorentz_cost(param, *args):

    eps_inf, w_p, gamma, f1, w1 = param
    
    hw, data = args
    
    model = cp.drude_lorentz_model(
        hw/hbar, 
        eps_inf=eps_inf, 
        w_p=w_p, 
        gamma=gamma,
        f_1 = f1,
        w_1 = w1)

    res = (model - data)*(
        hw**10.
#         >fit_cutoff
        )
    
    return np.array([res.real, res.imag]).ravel()
    
```

```python
ini_params = [au_eps_inf, au_w_p, au_gamma, 0, 1e16]

drude_fit_0K = opt.least_squares(
    drude_lorentz_cost,
    ini_params,
    args=(hw_eps_T_dict[f'hw_eps_0K'][0], hw_eps_T_dict[f'hw_eps_0K'][1]),
    x_scale=[1e2, 1e16, 1e14, 1, 1e16]
    )
```

```python
def idx_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx
```

### Plot contributions to linear model with thermal radius dependence

```python
def alpha_0(eps_c, eps_b, a, r_th, dndT_c):
    """ eps_c and dndT should be functions of frequency"""
    
    dnbdT = -2.7*10**(-4)
    
    alpha = (
        6*a**3. 
        *
        (eps_b/(eps_c+2*eps_b)**2.)
        *
        eps_c**0.5
        *
        dndT_c
        *
        (1/(np.pi*a))
        *
        ((r_th+a)/r_th)
        *
        (1/(((r_th+a)/r_th)**2.+(a/r_th)**2.))
        )
    return alpha
    
def alpha_pion2(eps_c, eps_b, a, r_th, dndT_c):
    """ eps_c and dndT should be functions of frequency"""
    
    dndT_b = -2.7*10**(-4)
    
    alpha = ((
        6*a**3. 
        *
        eps_b/(eps_c+2*eps_b)**2.
        *
        eps_c**0.5
        *
        dndT_c
        *
        (1/(np.pi*a))
        *
        ((r_th+a)/r_th)
        *
        (1/(((r_th+a)/r_th)**2.+(a/r_th)**2.))
        )
        +
        ((
            1
            + 
            ((2*a**3./r_th**3.)
            *
            (eps_c - eps_b)**2./(eps_c+2*eps_b)**2.))
        *
        1/eps_b**0.5
        *
        dndT_b
        *
        r_th**2.
        ))
    return alpha

def alpha_phi(eps_c, eps_b, a, r_th, dndT_c):
    """ eps_c and dndT should be functions of frequency"""
    
    dndT_b = -2.7*10**(-4)
    
    alpha = (
        (
            1+ (2*a**3./r_th**3.
            *
            (eps_c - eps_b)**2./(eps_c+2*eps_b)**2.))
        *
        (1/eps_b**0.5)
        *
        dndT_b
        *
        (-np.exp(-(r_th-a)/r_th)/(((r_th+a)/r_th)**2.+(a/r_th)**2.))
        *
        r_th*(r_th-a)
        )
    return alpha
    
def alpha_pion2phi(eps_c, eps_b, a, r_th, dndT_c):
    """ eps_c and dndT should be functions of frequency"""
    
    dndT_b = -2.7*10**(-4)
    
    alpha = (
        (
            1+ 2*a**3./r_th**3.
            *
            (eps_c - eps_b)**2./(eps_c+2*eps_b)**2.)
        *
        1/eps_b**0.5
        *
        dndT_b
        *
        (-np.exp(-(r_th-a)/r_th)/(((r_th+a)/r_th)**2.+(a/r_th)**2.))
        *
        r_th*(2*r_th+3*a)
        )
    return alpha
```
